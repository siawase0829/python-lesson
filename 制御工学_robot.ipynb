```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 伝達関数 G(s) = (s + 2) / (s^2 + 2s + 5)
numerator = [1, 2]  # 分子 (s + 2)
denominator = [1, 2, 5]  # 分母 (s^2 + 2s + 5)

# 伝達関数を定義
sys = signal.TransferFunction(numerator, denominator)

# 周波数応答を計算
w, mag, phase = signal.bode(sys)

# ボード線図 (周波数 vs 振幅)
plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(w, mag)  # 周波数軸を対数表示
plt.title('Bode plot')
plt.ylabel('Magnitude (dB)')
plt.grid(True)

plt.show()
# ボード線図 (周波数 vs 位相)
plt.subplot(2, 1, 2)
plt.semilogx(w, phase)
plt.ylabel('Phase (degrees)')
plt.xlabel('Frequency (rad/s)')
plt.grid(True)


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 伝達関数 G(s) = (s + 2) / (s^2 + 2s + 5)
numerator = [1, 2]  # 分子 (s + 2)
denominator = [1, 2, 5]  # 分母 (s^2 + 2s + 5)

# 伝達関数を定義
sys = signal.TransferFunction(numerator, denominator)

# 周波数範囲を設定
w = np.logspace(-2, 2, 1000)  # 0.01から100までの対数刻み

# 周波数応答を取得
w, h = signal.freqresp(sys, w=w)

# ナイキスト線図を描画 (実部 vs 虚部)
plt.plot(h.real, h.imag, label='Nyquist plot')
plt.plot(h.real, -h.imag, label='Conjugate', linestyle='--')  # 対称部分も表示

# 原点 (0, 0) にクロスマークを追加
plt.plot(0, 0, 'ro')

# グラフの設定
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Nyquist plot')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.legend()

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots()

# ブロックを描画
block = patches.FancyBboxPatch((0.2, 0.4), 0.5, 0.3, boxstyle="round,pad=0.05",
                               edgecolor='black', facecolor='lightgrey')
ax.add_patch(block)

# 矢印を描画
arrow = patches.FancyArrowPatch((0.2, 0.4), (0.2, 0.7), mutation_scale=20,
                                arrowstyle='->', color='black')
ax.add_patch(arrow)

# 軸の設定
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')  # 軸を非表示にする

plt.show()
```

```python
from graphviz import Digraph

dot = Digraph()

# ノード（ブロック）を追加
dot.node('A', 'ブロック A')
dot.node('B', 'ブロック B')

# エッジ（矢印）を追加
dot.edge('A', 'B', label='矢印')

# 出力ファイルを作成
dot.render('block_diagram', format='png', view=True)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# グラフを作成
G = nx.DiGraph()
G.add_node('A', label='ブロック A')
G.add_node('B', label='ブロック B')
G.add_edge('A', 'B', label='矢印')

# 描画
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightgrey', font_size=10, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['label'] for u, v, d in G.edges(data=True)})
plt.show()
```
